#### Description
This dataset contains 1.2M distinct tweets about the current ongoing Ukraine-Russia conflict.

Implementation
Two Jupyter notebooks running 24/7, executing every 15 mins monitoring hashtags pertaining to the ongoing Ukraine-Russia conflict. The dataset creator also implemented a simple "hashtag crawler" where the dataset creator crawled the top-most hashtags from an initial set of hashtags thus getting the other related hashtags at a given point in time.

```python
usa_tweets_dataset = en_tweets_df[(en_tweets_df['location'] == 'USA')|(en_tweets_df['location'] == 'United States')|(en_tweets_df['location'] == 'united states') | (en_tweets_df['location'] == 'US')| (en_tweets_df['location'] == 'usa')| (en_tweets_df['location'] == 'us')]
usa_tweets_dataset.head()
```

```python
# reducing columns to export only essential columns
reduced_usa_tweets_dataset = usa_tweets_dataset.drop(columns=["Unnamed: 0","username","acctdesc","usercreatedts","tweetcreatedts","userid","hashtags","language","coordinates","hashtags","extractedts"])
reduced_usa_tweets_dataset.head()
```

```python
reduced_usa_tweets_dataset['text'] = reduced_usa_tweets_dataset['text'].str.replace(',', '')
reduced_usa_tweets_dataset['text'] = reduced_usa_tweets_dataset['text'].str.replace('#', '')
reduced_usa_tweets_dataset
```

```python
# extract the USA dataframe into a csv file
reduced_usa_tweets_dataset.to_csv('usa_tweets_dataset_cleaned.csv', index=False)
```
